```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('diabetes.csv')
df.shape
```

```python
df.head()
```

```python
df.info()  #결측치 없네 개꿀
```

```python
df.shape, df.shape[0]
```

```python
split_data=int(df.shape[0]*0.8)
split_data
```

```python
train=df[:split_data].copy()
train
```

```python
test=df[split_data:].copy()
test
```

```python
#학습과 예측에 사용할 피쳐컬럼 나누기
train.columns
```

```python
feature_names=train.columns[:-1].tolist()
feature_names
```

```python
label_name=train.columns[-1]
label_name
```

```python
x_train=train[feature_names]
x_train.head() ,x_train.shape
```

```python
y_train=train[label_name]
y_train.head() , y_train.shape
```

```python
x_test=test[feature_names]
x_test.head() ,x_test.shape
```

```python
y_test=test[label_name]
y_test.head() , y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier

model=DecisionTreeClassifier()
model
```

```python
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
y_predict
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(20,20))
tree=plot_tree(model,feature_names=feature_names,filled=True,fontsize=10)
```

```python
#어떤 feature가 중요한가... 유용한 기능인듯

model.feature_importances_
```

```python
sns.barplot(x=model.feature_importances_,y=feature_names)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_predict)
```
